```python
from sympy import *
init_printing()
import numpy as np
from matplotlib import pyplot as plt
```

```python
a,b,e0,h,i,t,e,g1,g2,g3 = symbols("A, B, E_0,h,i,t,e,g_1,g_2,g_3")
```

```python
H = Matrix([[e0,-a,0],[-a,e0,-b],[0,-b,e0]])
display(H)
```

```python
H.eigenvals()
```

```python
H.eigenvects()
```

```python
P,D = H.diagonalize() #diagonaling H expressing as diagonal  matrix
```

```python
P #basis transformation which takes from eigen basis to physical basis
```

```python
D #eigen mattrix in terms of eigenvalues
```

```python
simplify(P.inv()) #inverse of change of basis matrix
```

```python
U = Matrix([[e**g1,0,0],[0,e**g2,0],[0,0,e**g3]]) #matrix of solutions in the eigen-basis
```

```python
U
```

```python
P*U
```

```python
f = simplify(P*U*P.inv())
```

```python
f
```

```python
W = Matrix([[1], [0], [0]])
```

```python
W #initial conditions in physical basis
```

```python
f*W #returned solutions to physical basis
```

```python
#Probability densities and their variation

def psi_e(kt):
  return (0.5*np.cos(kt) + 0.5)**2

def psi_mu(kt):
  return (np.sin(kt))**2/2

def psi_tau(kt):
  return (0.5*np.cos(kt) - 0.5)**2

kt = np.arange(0, 3*np.pi, 0.01)

plt.plot(kt, psi_e(kt), c = "indigo",label = "Electron")
plt.plot(kt, psi_mu(kt), c = "blue",label = "Muon")
plt.plot(kt, psi_tau(kt), c = "black",label = "Tau")
plt.xlabel("kt")
plt.ylabel("Probability Density" )
plt.legend()
plt.show()
```
